```python
plt.scatter(iris.SepalLengthCm, iris.SepalWidthCm)
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Iris Dataset')
plt.show()
```

```python
plt.bar(iris.Species, iris.SepalLengthCm)
plt.title('Bar plot of Species and Sepal Length (cm)')
plt.xlabel('Species')
plt.ylabel('Sepal Length (cm)')
plt.show()
```

```python
plt.hist(iris.SepalWidthCm)
plt.title('Histogram showing Species and Sepal Width')
plt.xlabel('Species')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
plt.plot(iris.PetalLengthCm, iris.PetalWidthCm)
plt.scatter(iris.PetalLengthCm, iris.PetalWidthCm)
plt.title('Line plot of Petal length and Petal Width')
plt.xlabel('Petal Lenght (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
plt.plot(range(0,100),iris.shape[0]), iris.PetalWidthCm)
#plt.scatter(iris.PetalLengthCm, iris.PetalWidthCm)
plt.title('Line plot of Petal length and Petal Width')
plt.xlabel('Petal Lenght (cm)')
plt.ylabel('Petal Width (cm)')
plt.show()
```

```python
columns = iris.drop(['Species', 'Id'], axis =1)
for column in columns:
    plt.plot(range(0,iris.shape[0]), iris[column])
    #plt.scatter(range(0, iris.shape[0]), iris[column])
#plt.scatter(iris.PetalLengthCm, iris.PetalWidthCm)
plt.title('Line plot of Petal length and Petal Width')
plt.xlabel('Observation')
plt.ylabel('Features (cm)')
plt.legend()
plt.show()
```

```python
# plot a line, implicitly creating a subplot(111)
plt.plot([1,2,3])
# now create a subplot which represents the top plot of a grid
# with 2 rows and 1 column. Since this subplot will overlap the
# first, the plot (and its axes) previously created, will be removed
plt.subplot(211)
plt.plot(range(12))
plt.subplot(212, facecolor='y') # creates 2nd subplot with yellow background
```
